#### Converting GWh to MWh

```python
Country_sheet_11_11['Electricity Generation (MWh)'] = Country_sheet_11_11['Electricity Generation (GWh)'] * 1000
```

```python
## Define Emission Factors

emission_factors = {
    'Natural gas': 490,
    'Oil': 740,
    'Coal and peat': 935,  # Average of 820-1050
    'Fossil fuels n.e.s.': 800,
    'Other primary solid biofuels n.e.s.': 275,  # Average of 100-450
    'Non-renewable municipal waste': 750,  # Average of 500-1000
    'Nuclear': 0,
    'Other non-renewable energy': 850  # Average of 700-1000
}
```

```python
## Calculate Carbon Emissions

Country_sheet_11_11['Carbon Emission (kg CO₂)'] = Country_sheet_11_11.apply(
    lambda row: (
        row['Electricity Generation (MWh)'] * emission_factors.get(row['Sub-Technology'], 0)
        if row['Electricity Generation (GWh)'] > 0 else 0
    ),
    axis=1
)
```
